```python
from mtranslate import translate

def back_translate(sentence):
    # Translate from English to Korean
    korean_translation = translate(sentence, 'ko', 'en')

    # Translate from Korean to Arabic
    arabic_translation = translate(korean_translation, 'ar', 'ko')

    # Translate from Arabic to French
    french_translation = translate(arabic_translation, 'fr', 'ar')

    # Translate from French back to English
    back_english_translation = translate(french_translation, 'en', 'fr')

    return back_english_translation
```
